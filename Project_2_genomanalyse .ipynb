```python
import lightkurve as lk
data=lk.search_targetpixelfile("Kepler-702", author="Kepler", quarter=4, cadence="long").download() #trying to get the same dataset as before, quater, cadence are to filter results
lc = data.to_lightcurve(method="pld").remove_outliers().flatten() #transforming the data into a lightcurve, optimising data for plotting
period=lc.to_periodogram("bls").period_at_max_power #lightcurve documentation to get the period 
lc.fold(period).scatter() #print the values for the time of the period
```
